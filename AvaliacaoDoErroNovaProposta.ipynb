```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_vector(size, min_val, max_val):
    return np.random.uniform(min_val, max_val, size)

def calculate_mean(vector):
    return np.mean(vector)

def calculate_std(vector):
    return np.std(vector, ddof=1)

def calculate_covariance(a, b):
    return np.cov(a, b, ddof=1)[0][1]

def calculate_f_b(a, b):
    std_a = calculate_std(a)
    covariance_ab = calculate_covariance(a, b)
    f_b = covariance_ab / std_a
    return f_b

# Parmetros para gerao de vetores aleatrios
min_val = 1
max_val = 10
num_samples = 10

# Listas para armazenar resultados
sizes = []
f_b_values = []
variances_a = []

for _ in range(num_samples):
    size = np.random.randint(5, 1000)  # Tamanho aleatrio entre 5 e 15
    a = generate_random_vector(size, min_val, max_val)
    b = generate_random_vector(size, min_val, max_val)

    f_b_value = calculate_f_b(a, b)
    variance_a = calculate_std(a) ** 2

    sizes.append(size)
    f_b_values.append(f_b_value)
    variances_a.append(variance_a)

# Plotando f(b) em relao ao tamanho do vetor
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.plot(sizes, f_b_values, 'o-', label='f(b)')
plt.xlabel('Tamanho do Vetor')
plt.ylabel('f(b)')
plt.title('f(b) vs. Tamanho do Vetor')
plt.legend()

# Plotando varincia de a em relao ao tamanho do vetor
plt.subplot(1, 3, 2)
plt.plot(sizes, variances_a, 's-', label='Varincia de a', color='orange')
plt.xlabel('Tamanho do Vetor')
plt.ylabel('Varincia de a')
plt.title('Varincia de a vs. Tamanho do Vetor')
plt.legend()

# Plotando f(b) em relao  varincia de a
plt.subplot(1, 3, 3)
plt.plot(variances_a, f_b_values, 'd-', label='f(b) vs. Varincia de a', color='green')
plt.xlabel('Varincia de a')
plt.ylabel('f(b)')
plt.title('f(b) vs. Varincia de a')
plt.legend()

plt.tight_layout()
plt.show()
```
