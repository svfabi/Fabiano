```python
import numpy as np
import matplotlib.pyplot as plt

# Função para calcular k
def calcular_k(x, y):

    n = len(x)

    eqc1 =  ((0.01*n)**n )
    eqc2 = n

    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_xy = np.sum(x * y)
    k = (sum_x * sum_y) / (eqc2* sum_xy)

    return k, sum_x, sum_y

# Gerar vetores aleatórios e calcular k
n_vetores = 100
comprimentos = np.random.randint(2, 50, n_vetores)
ks = []
somas_x = []
somas_y = []

for n in comprimentos:
    x = np.random.randint(0, 500, n)
    y = np.random.randint(0, 500, n)
    k, sum_x, sum_y = calcular_k(x, y)
    ks.append(k)
    somas_x.append(sum_x)
    somas_y.append(sum_y)

# Plotar os resultados
plt.figure(figsize=(12, 8))

# Gráfico de k
plt.subplot(3, 1, 1)
plt.scatter(comprimentos, ks, color='blue', alpha=0.6)
plt.xlabel('Comprimento do Vetor (n)')
plt.ylabel('Valor de k')
plt.title('Valores de k para Vetores Aleatórios')
plt.grid(True)

# Gráfico da soma dos valores de x
plt.subplot(3, 1, 2)
plt.scatter(comprimentos, somas_x, color='green', alpha=0.6)
plt.xlabel('Comprimento do Vetor (n)')
plt.ylabel('Soma dos Valores de x')
plt.title('Soma dos Valores de x para Vetores Aleatórios')
plt.grid(True)

# Gráfico da soma dos valores de y
plt.subplot(3, 1, 3)
plt.scatter(comprimentos, somas_y, color='red', alpha=0.6)
plt.xlabel('Comprimento do Vetor (n)')
plt.ylabel('Soma dos Valores de y')
plt.title('Soma dos Valores de y para Vetores Aleatórios')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Função para calcular k com normalização em função de n
def calcular_k(x, y):
    n = len(x)
    eqc1 = ((0.01 * n) ** n)
    eqc2 = n
    eqc3 = 1 / np.sqrt(n)  # Novo fator de normalização em função de n

    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_xy = np.sum(x * y)
    k = (sum_x * sum_y) / (eqc3 * sum_xy * eqc3)  # Ajuste na fórmula de k

    return k, sum_x, sum_y

# Gerar vetores aleatórios e calcular k
n_vetores = 100
comprimentos = np.random.randint(2, 50, n_vetores)
ks = []
somas_x = []
somas_y = []

for n in comprimentos:
    x = np.random.randint(0, 500, n)
    y = np.random.randint(0, 500, n)
    k, sum_x, sum_y = calcular_k(x, y)
    ks.append(k)
    somas_x.append(sum_x)
    somas_y.append(sum_y)

# Plotar os resultados
plt.figure(figsize=(12, 8))

# Gráfico de k
plt.subplot(3, 1, 1)
plt.scatter(comprimentos, ks, color='blue', alpha=0.6)
plt.xlabel('Comprimento do Vetor (n)')
plt.ylabel('Valor de k')
plt.title('Valores de k para Vetores Aleatórios')
plt.grid(True)
plt.show()
```
