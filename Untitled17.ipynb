```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score

# Dados fornecidos
x = np.array([1.00000000e-03, 2.06913808e-03, 4.28133240e-03, 8.85866790e-03, 1.83298071e-02, 3.79269019e-02, 7.84759970e-02, 1.62377674e-01,
              3.35981829e-01, 6.95192796e-01, 1.43844989e+00, 2.97635144e+00, 6.15848211e+00, 1.27427499e+01, 2.63665090e+01, 5.45559478e+01, 1.12883789e+02, 2.33572147e+02, 4.83293024e+02, 1.00000000e+03])
y = np.array([0.500999, 0.50206487, 0.50426308, 0.50878088, 0.51799987, 0.53654101, 0.57276564, 0.63969442, 0.75148683, 0.9100966, 1.0899034,
              1.24851317, 1.36030558, 1.42723436, 1.46345899, 1.48200013, 1.49121912, 1.49573692, 1.49793513, 1.499001])

# Função combinada exponencial e logarítmica sem os parâmetros c1 e b2
def combined_func(x, a1, b1, a2):
    return np.piecewise(x, [x <= 1, x > 1], [lambda x: a1 * np.exp(b1 * x), lambda x: a2 * np.log(x)])

# Fornecendo um palpite inicial para os parâmetros
p0 = [1, 1, 1]  # Palpite inicial para a1, b1 e a2

# Ajustando a curva combinada aos dados
params_combined, covariance_combined = curve_fit(combined_func, x, y, p0=p0, maxfev=5000)

# Parâmetros ajustados para a função combinada
a1_combined, b1_combined, a2_combined = params_combined

# Gerando valores ajustados para plotagem
x_fit = np.linspace(min(x), max(x), 100)
y_fit_combined = combined_func(x_fit, a1_combined, b1_combined, a2_combined)

# Calculando o R²
y_pred_combined = combined_func(x, a1_combined, b1_combined, a2_combined)
r2_combined = r2_score(y, y_pred_combined)

# Plotando os dados originais e a curva ajustada
plt.scatter(x, y, label='Dados Originais')
plt.plot(x_fit, y_fit_combined, label=f'Curva Combinada Ajustada (R²={r2_combined:.4f})', color='red')
plt.xscale('log')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Ajuste da Curva Combinada Exponencial e Logarítmica')
plt.ylim(0.5, 1.5)  # Definindo os limites do eixo y
plt.show()

# Exibindo os parâmetros ajustados e o R²
print(f"Parâmetros ajustados para a função combinada: a1 = {a1_combined}, b1 = {b1_combined}, a2 = {a2_combined}")
print(f"Coeficiente de determinação (R²): {r2_combined:.4f}")
```
